```python
def list_up(list, n):
	while n > 0:
		elem = list.pop(0)
		list.append(elem)
		n -= 1

def list_down(list, n):
	while n > 0:
		elem = list.pop(len(list) - 1)
		list.insert(0, elem)
		n -= 1
```
